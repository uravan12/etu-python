```python
import pandas as pd
```

```python
df = pd.read_csv('adult/adult.data', header = None, names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'])
df.head(20)
```

```python
# Число столбцов в наборе данных? 
num_colums = df.shape[1]
print(f"Количество столбцов: {num_colums}")
```

```python
# Есть ли пропуски в данных? Если есть, то в каких столбцах

missing_data = {}

for column in df.columns:
    num_missing = (df[column] == ' ?').sum()
    
    if num_missing > 0:
        missing_data[column] = num_missing

if missing_data:
    print("Столбцы, содержащие пропуски :")
    for column, count in missing_data.items():
        print(f"{column}: {count} пропущенных значений")
else:
    print("Пропуски отсутствуют во всех столбцах.")
```

```python
# Кол-во уникальных значений в столбце race
num_unique_races = df['race'].nunique()

print(f"Количество уникальных значений в столбце 'race': {num_unique_races}")
```

```python
# Медиана hours-per-week
median_hours_per_week = df['hours-per-week'].median()

print(f"Медиана hours-per-week: {median_hours_per_week}")
```

```python
# Кого больше - женщин или мужчин с ЗП >50K?

high_earners = df[df['income'] == ' >50K'] 


num_males = (high_earners['sex'] == ' Male').sum()
num_females = (high_earners['sex'] == ' Female').sum()


if num_males > num_females:
    print("Мужчин с ЗП > 50K больше, чем женщин.")
elif num_females > num_males:
    print("Женщин с ЗП > 50K больше, чем мужчин.")
else:
    print("Количество мужчин и женщин с ЗП > 50K одинаково.")

print(f"Мужчин: {num_males}, Женщин: {num_females}")
```

```python
# Заполните пропущенные данные в отдельных столбцах наиболее встречаемыми значениями. Как еще можно было бы заполнить пропущенные данные?
columns = ['workclass', 'occupation', 'native-country']
for col in columns:
    most_common = df[col].mode()[0]
    df.replace({col: ' ?'}, most_common, inplace=True)

```
